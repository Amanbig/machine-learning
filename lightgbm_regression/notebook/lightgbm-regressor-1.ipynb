# **Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **Importing dataset**

```python
dataset = pd.read_csv('/kaggle/input/student-performance-multiple-linear-regression/Student_Performance.csv')
dataset
```

# **Applying Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['Extracurricular Activities'] = le.fit_transform(dataset['Extracurricular Activities'])
dataset
```

# **Applying Standard Scaler**

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
dataset = ss.fit_transform(dataset)
dataset
```

# **Splitting into X and Y**

```python
X = dataset[:,:-1]
Y = dataset[:,-1]
```

# **Splitting into test and train set**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25)
```

# **Fitting model Lightgbm**

```python
from lightgbm import LGBMRegressor
lg = LGBMRegressor()
lg.fit(X_train,Y_train)
```

# **Checking error of prediction**

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
Y_pred = lg.predict(X_test)
ma = mean_absolute_error(Y_pred,Y_test)
ms = mean_squared_error(Y_pred,Y_test)
print(ma)
print(ms)
```

# **Saving model**

```python
import pickle

with open("model.pkl", "wb") as f:
    pickle.dump(lg,f)
```
