# **Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **Importing Dataset**

```python
dataset_train = pd.read_csv('/kaggle/input/random-linear-regression/train.csv')
dataset_test = pd.read_csv('/kaggle/input/random-linear-regression/test.csv')
```

# **Detecting Null values**

```python
dataset_train.isnull()
```

```python
dataset_train.describe()
```

# **Dropping NAN Values**

```python
dataset_train = dataset_train.dropna()
dataset_test = dataset_test.dropna()
```

# **Splitting into dependent and independent variable**

```python
X_train = dataset_train.iloc[:,:-1]
Y_train = dataset_train.iloc[:,-1]
X_test = dataset_test.iloc[:,:-1]
Y_test = dataset_test.iloc[:,-1]
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)
```

# **Predicting Test Set**

```python
Y_pred = lr.predict(X_test)
```

# **Checking Mean Square**

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_pred,Y_test)
print(mse)
```
