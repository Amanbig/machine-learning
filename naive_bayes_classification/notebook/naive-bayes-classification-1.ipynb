# **Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **Importing dataset**

```python
dataset = pd.read_csv('/kaggle/input/glass/glass.csv')
dataset
```

# **Splitting into X, Y**

```python
X = dataset.iloc[:,:-1]
Y = dataset.iloc[:,-1]
```

# **Splitting into train and test set**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train, Y_test = train_test_split(X,Y,test_size=0.25,random_state=42)
```

# **Applying gaussian and multinomial**

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB
gn = GaussianNB()
mn = MultinomialNB()
gn.fit(X_train,Y_train)
mn.fit(X_train,Y_train)
```

# **Checking errors**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
Y_pred = gn.predict(X_test)
Y_pred1 = mn.predict(X_test)
print("Gaussian NB")
print(mean_absolute_error(Y_pred,Y_test))
print(mean_squared_error(Y_pred,Y_test))
print("Multinomial NB")
print(mean_absolute_error(Y_pred1,Y_test))
print(mean_squared_error(Y_pred1,Y_test))
```

# **Saving model**

```python
import pickle

with open("model_gaussian.pkl","wb") as f:
    pickle.dump(gn,f)

with open("model_multinomial", "wb") as f:
    pickle.dump(mn,f)
```
