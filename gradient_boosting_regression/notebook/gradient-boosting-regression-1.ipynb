# **Importing Libaries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **Importing dataset**

```python
dataset = pd.read_csv('/kaggle/input/home-value-insights/house_price_regression_dataset.csv')
dataset
```

# **Applying Standard Scaling**

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
dataset = ss.fit_transform(dataset)
```

# **Spliting into value to be predicted and given value**

```python
X = dataset[:,:-1]
Y = dataset[:,-1]
```

# **Splitting into test and train set**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25)
```

# **Applying Gradient Boosting**

```python
from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor()
gb.fit(X_train,Y_train)
```

# **Training model using hist Gradient Boosting**

```python
from sklearn.ensemble import HistGradientBoostingRegressor
hg = HistGradientBoostingRegressor()
hg.fit(X_train,Y_train)
```

# **Checking Scores of each**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
Y_pred = gb.predict(X_test)
Y_pred_hist = hg.predict(X_test)

ma = mean_absolute_error(Y_pred,Y_test)
ms = mean_squared_error(Y_pred,Y_test)

ma_hist = mean_absolute_error(Y_pred_hist,Y_test)
ms_hist = mean_squared_error(Y_pred_hist, Y_test)

print('Gradient boosting:')
print(ma)
print(ms)

print('Hist Gradient Boosting')
print(ma_hist)
print(ms_hist)
```

# **Saving model**

```python
import pickle

with open("model.pkl","wb") as f:
    pickle.dump(gb,f)

with open("model_hist.pkl","wb") as f:
    pickle.dump(hg,f)
```
