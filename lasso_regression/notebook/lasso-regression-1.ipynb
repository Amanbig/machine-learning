# **Importing Library**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **Importing Dataset**

```python
dataset = pd.read_csv("/kaggle/input/usa-housing-dataset/USA_Housing.csv")
dataset
```

# **Droping address column**

```python
dataset.drop('Address',axis=1,inplace=True)
```

# **Applying standard scaler**

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
dataset = ss.fit_transform(dataset)
```

# **Splitting into dependent and independent data**

```python
X = dataset[:,:-1]
Y = dataset[:,-1]
```

# **Splitting into train and test set**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25)
```

# **Training Lasso Model**

```python
from sklearn.linear_model import Lasso
l = Lasso()
l.fit(X_train,Y_train)
```

# **Calculating errors**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
Y_pred = l.predict(X_test)
ms = mean_squared_error(Y_pred,Y_test)
ma = mean_absolute_error(Y_pred,Y_test)
print(ms)
print(ma)
```

# **Saving Model**

```python
import pickle

with open("model.pkl","wb") as f:
    pickle.dump(l,f)
```
