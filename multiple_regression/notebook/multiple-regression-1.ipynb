# **Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **Importing dataset**

```python
dataset = pd.read_csv('/kaggle/input/student-performance-multiple-linear-regression/Student_Performance.csv')
dataset.head()
```

```python
dataset['Extracurricular Activities'] = dataset['Extracurricular Activities'].map({"Yes":1, "No":0})
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
dataset = ss.fit_transform(dataset)
```

```python
dataset
```

```python
X = dataset[:,:-1]
Y = dataset[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)
```

```python
lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
ms = mean_squared_error(lr.predict(X_test),Y_test)
mb = mean_absolute_error(lr.predict(X_test),Y_test)
print(ms)
print(mb)
```

```python
import pickle

with open("model.pkl","wb") as f:
    pickle.dump(lr,f)
```
