# **Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# **Importing Dataset**

```python
dataset = pd.read_csv('/kaggle/input/home-value-insights/house_price_regression_dataset.csv')
dataset
```

# **Applying Standard Scaler**

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
dataset = ss.fit_transform(dataset)
```

# **Splitting into given value and value to be predicted**

```python
X = dataset[:,:-1]
Y = dataset[:,-1]
```

# **Splitting into test and train set**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25)
```

# **Training ElasticNet**

```python
from sklearn.linear_model import ElasticNet
en = ElasticNet()
en.fit(X_train,Y_train)
```

# **Checking model errors**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
Y_pred = en.predict(X_test)
ma = mean_absolute_error(Y_pred,Y_test)
ms =mean_squared_error(Y_pred,Y_test)
print(ma)
print(ms)
```

# **Saving model**

```python
import pickle

with open("model.pkl","wb") as f:
    pickle.dump(en,f)
```
