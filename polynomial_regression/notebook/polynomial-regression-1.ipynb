# **Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **Importing Dataset**

```python
dataset = pd.read_csv('/kaggle/input/polynomial-regression/Ice_cream selling data.csv')
dataset.head()
```

# **Applying Standard Scaler**

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
dataset = ss.fit_transform(dataset)
```

# **Splitting values for dependent and independent variables**

```python
X = dataset[:,:-1]
Y = dataset[:,-1]
```

# **Finding the polynomial Features**

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=2)
X = pf.fit_transform(X)
```

# **Splitting into train and test set**

```python
from sklearn.model_selection import train_test_split
X_train,X_test, Y_train,Y_test = train_test_split(X,Y)
```

# **Now applying Linear Regression**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)
```

# **Checking errors**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
ma = mean_absolute_error(lr.predict(X_test),Y_test)
ms = mean_absolute_error(lr.predict(X_test),Y_test)
print(ma)
print(ms)
```

# **Saving the model**

```python
import pickle

with open("model.pkl","wb") as f:
    pickle.dump(lr,f)
```
