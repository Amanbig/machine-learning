# **Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **Importing dataset**

```python
dataset = pd.read_csv('/kaggle/input/home-value-insights/house_price_regression_dataset.csv')
dataset
```

# **Applying Standard Scaling**

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
dataset = ss.fit_transform(dataset)
```

# **Splitting into dependent and independent variable**

```python
X = dataset[:,:-1]
Y = dataset[:,-1]
```

# **Splitting into train and test set**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=42)
```

# **Applying random Forest**

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_train,Y_train)
```

# **Checking errors**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
Y_pred = rf.predict(X_test)
ma = mean_absolute_error(Y_pred,Y_test)
ms = mean_squared_error(Y_pred,Y_test)
print(ma)
print(ms)
```

# **Saving model**

```python
import pickle

with open("model.pkl","wb") as f:
    pickle.dump(rf,f)
```
