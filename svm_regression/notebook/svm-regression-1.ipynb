# **Importing Librarys**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **Importing Dataset**

```python
dataset = pd.read_csv('/kaggle/input/student-performance-multiple-linear-regression/Student_Performance.csv')
dataset
```

# **Applying Label Encoder to Extracurricular Activies column**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['Extracurricular Activities'] = le.fit_transform(dataset['Extracurricular Activities'])
dataset
```

# **Splitting into dependent and independent variables**

```python
X = dataset.iloc[:,:-1]
Y = dataset.iloc[:,-1]
```

# **Splitting into train and test set**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25)
```

# **Applying support vector regression**

```python
from sklearn.svm import SVR
svc = SVR(kernel = 'rbf')
svc.fit(X_train,Y_train)
```

# **Checking error in prediction**

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
Y_pred = svc.predict(X_test)
ma = mean_absolute_error(Y_pred,Y_test)
ms = mean_squared_error(Y_pred,Y_test)
print(ma)
print(ms)
```

# **Saving Model**

```python
import pickle

with open("model.pkl","wb") as f:
    pickle.dump(svc,f)
```
