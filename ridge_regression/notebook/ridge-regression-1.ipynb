# **Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **Importing Dataset**

```python
dataset = pd.read_csv('/kaggle/input/algerian-forest-fires-cleaned-dataset/Algerian_forest_fires_cleaned_dataset.csv')
dataset
```

# **Dividing dataset into one which we need to find and one with which other is to be find**

```python
X = dataset.iloc[:,:-2]
Y = dataset.iloc[:,-2]
```

# **Applying LabelEncoder**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(Y)
```

# **Splitting into train and test**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25)
```

# **Applying Ridge Regression**

```python
from sklearn.linear_model import Ridge
rr = Ridge()
rr.fit(X_train,Y_train)
```

# **Getting the error in result by model**

```python
Y_pred = rr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error
ma = mean_absolute_error(Y_pred,Y_test)
ms = mean_squared_error(Y_pred,Y_test)
print(ms)
print(ma)
```
